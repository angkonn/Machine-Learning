
**Unsupervised Machine Learning**  
**Objective:** Discover hidden patterns, clusters, or relationships in **unlabeled data** (data without predefined labels or categories).  

---

### **Key Concepts**  
1. **Clustering**  
   - Groups data points based on similarity.  
   - **Common Algorithms:**  
     - **K-Means:** Divides data into `n_clusters` groups.  
       - **Formula:** Minimize the sum of squared distances:  
         \[ \sum_{i=1}^{n} (x_i - \mu_j)^2 \]  
         where \(\mu_j\) = cluster center.  
     - **DBSCAN:** Identifies clusters based on data density (handles noise).  
     - **Hierarchical Clustering:** Creates a tree-like structure (dendrogram).  

   ```python
   # K-Means Example
   from sklearn.cluster import KMeans
   kmeans = KMeans(n_clusters=3)
   clusters = kmeans.fit_predict(X)
   ```

2. **Dimensionality Reduction**  
   - Reduces features while retaining critical information.  
   - **Common Algorithms:**  
     - **PCA (Principal Component Analysis):**  
       - Uses eigenvectors to maximize variance:  
         \[ X = U \Sigma V^T \]  
       - **Explained Variance Ratio:** Shows how much information each component retains.  
     - **t-SNE:** Visualizes high-dimensional data in 2D/3D.  

   ```python
   # PCA Example
   from sklearn.decomposition import PCA
   pca = PCA(n_components=2)
   X_pca = pca.fit_transform(X)
   ```

3. **Association Rule Learning**  
   - Finds relationships between items (e.g., "customers who buy X also buy Y").  
   - **Common Algorithm:**  
     - **Apriori:** Measures rules using **Support**, **Confidence**, and **Lift**.  
       - Example: {Milk} → {Eggs} (probability of co-occurrence).  

---

### **Evaluation Metrics**  
- **Clustering:**  
  - **Silhouette Score (-1 to 1):** Higher = better-defined clusters.  
  - **Davies-Bouldin Index:** Lower = better clustering.  
- **Dimensionality Reduction:**  
  - **Explained Variance Ratio (PCA):** % of variance retained per component.

2. **Anomaly Detection:** Identify fraud or outliers.  
3. **Image Compression:** Reduce image size using PCA.  
4. **Recommendation Systems:** Find product associations (e.g., Amazon’s "Frequently Bought Together").

### **Clustering**  
Clustering is an **unsupervised learning** method used to group unlabeled data points into clusters based on their similarities. The goal is to ensure data points in the same cluster are more alike to each other than to those in other clusters. It helps discover hidden patterns, segment datasets, or simplify complex data.  

---

### **K-Means Clustering**  
**K-Means** is the most widely used clustering algorithm. It partitions data into **K clusters** (user-defined number), where each cluster is represented by its **centroid** (the mean of all points in the cluster).  

---

### **How K-Means Works**  
1. **Choose K**: Decide the number of clusters (e.g., K=3).  
2. **Initialize Centroids**: Randomly pick K data points as initial centroids.  
3. **Assign Points to Clusters**:  
   - Calculate the distance (usually **Euclidean**) between each data point and all centroids.  
   - Assign each point to the **nearest centroid**, forming K clusters.  
4. **Update Centroids**:  
   - Recalculate the centroid of each cluster as the mean of all points in that cluster.  
5. **Repeat**:  
   - Reassign points to the updated centroids and recalculate centroids until they stabilize (no further changes).  

---

### **Key Mathematical Objective**  
K-Means aims to minimize the **sum of squared distances** between data points and their cluster centroids:  
$\text{Inertia}$= $\sum_{i=1}^{n} (x_i - \mu_j)^2
$
- $x_i$: Data point  
- $mu_j$: Centroid of the cluster to which $x_i$ belongs  

---

### **Choosing the Right K (Number of Clusters)**  
Use the **Elbow Method**:  
1. Run K-Means for different values of K (e.g., K=1 to K=10).  
2. Plot the **inertia** (sum of squared distances) against K.  
3. The "elbow" point (where inertia starts decreasing slowly) is the optimal K.  

---

### **Strengths of K-Means**  
- Simple and fast for small-to-medium datasets.  
- Works well with **spherical** or evenly sized clusters.  
- Scales efficiently to large datasets.  

---

### **Limitations of K-Means**  
- Requires predefining K (number of clusters).  
- Sensitive to initial centroid positions (may converge to local optima).  
- Struggles with:  
  - Non-spherical or irregularly shaped clusters.  
  - Clusters of varying sizes or densities.  
  - Outliers/noisy data.  

---

### **Real-World Applications**  
1. **Customer Segmentation**: Group users by purchase history or behavior.  
2. **Image Compression**: Reduce color space by clustering similar pixel values.  
3. **Document Clustering**: Organize text data (e.g., news articles) by topic.  
4. **Anomaly Detection**: Identify outliers as data points far from all centroids.  

---

### **When to Use K-Means?**  
- You have unlabeled data and want to explore natural groupings.  
- The data has **globular clusters** (not elongated or irregular).  
- Speed and simplicity are priorities.  

---

### **Key Takeaways**  
- K-Means is a **centroid-based** algorithm.  
- Always **scale your data** (e.g., using standardization) before applying K-Means.  
- Validate clusters using domain knowledge or metrics like the **Silhouette Score**.  
- For complex cluster shapes, consider algorithms like **DBSCAN** or **Hierarchical Clustering**.



**Optimizing K-Means Clustering Using the WCSS Method**  

K-means clustering groups data points into "K" clusters. Choosing the right **K** is critical. The **Within-Cluster Sum of Squares (WCSS)** method helps find the best K.  

### What is WCSS?  
- **WCSS** measures how tightly points are packed within clusters.  
- It calculates the sum of squared distances between each point and the **center (centroid)** of its cluster.  
- *Lower WCSS* = tighter clusters.  

### How to Optimize K Using WCSS?  
1. **Try different K values**: Run K-means for K=1,2,3,... up to a reasonable max.  
2. **Calculate WCSS** for each K.  
3. **Plot the WCSS values**:  
   - X-axis = K values.  
   - Y-axis = WCSS.  
4. **Look for the "Elbow Point"**:  
   - The point where WCSS stops dropping sharply and starts to flatten (like a bent elbow).  
   - This K is optimal: adding more clusters doesn’t improve clustering much.  

### Why Does This Work?  
- Small K (too few clusters): High WCSS (clusters are too spread out).  
- Very large K (too many clusters): WCSS becomes very low, but clusters may overfit the data (no meaningful groups).  
- The "elbow" balances simplicity and accuracy.  

### Example:  
If WCSS drops a lot from K=2 to K=3, but only a little from K=3 to K=4, K=3 is likely optimal.  

**Note**: If the elbow isn’t clear, try other methods (like the Silhouette Score).

````python
# Create an empty list to store WCSS values for different K values
wcss = []

# Define the range of cluster numbers to test (from 1 to 14)
clusture_range = range(1,15)

# Loop through each cluster number (K) in the range
for k in clusture_range:
    # Create a K-means model with current K value (number of clusters)
    km2 = KMeans(n_clusters=k)

    # Fit the model to the data (using 'age' and 'score' columns)
    km2.fit(df[['age','score']])

    # Append the WCSS (inertia) of the model to the list
    # Inertia = sum of squared distances of points to their cluster center
    wcss.append(km2.inertia_)
````

````python
wcss
````

````python
plt.plot(clusture_range,wcss, marker = 'o')
plt.xlabel('K')
plt.ylabel('WCSS')
plt.show()
````
