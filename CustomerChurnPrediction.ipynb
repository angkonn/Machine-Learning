# Customer Churn Prediction
**Understanding Churn:**

“Churn” means customers stopping their service or cancelling their subscription.
The goal is to identify those customers before they leave.

**Data Collection:**
Historical data about customers is collected. This can include usage patterns, demographics, service interactions, and more.
Each customer record has a label indicating whether they have churned (left) or stayed.

**Building a Model:**
The project treats churn prediction as a classification problem.
Machine learning algorithms (like logistic regression, decision trees, or XGBoost) are used to learn patterns from the data that indicate a high likelihood of churn.

**Model Training & Evaluation:** The dataset is split into training and testing parts. The model is trained on the training set and then evaluated on the testing set.
Evaluation metrics (like accuracy, precision, recall) are used to measure how well the model predicts churn.

**Actionable Insights:** Once the model identifies customers who are likely to churn, the company can target these customers with retention strategies such as special offers, personalized communication, or improved service.

# Preprocessing Data

```python
#Converting categorial values into numerical ones
#label encoding is used here
# customer id is not necessary here
```

```python
df=df.drop('customerID',axis=1)
df
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

for column in df.columns:
    # If the column is numeric, skip it.
    if np.issubdtype(df[column].dtype, np.number):
        continue
    # Otherwise, encode the categorical column.
    df[column] = LabelEncoder().fit_transform(df[column])
df
```

```python
x=df.drop('Churn',axis=1)
y=df['Churn']
```

# Scaling the dataset
Scaling a dataset means transforming the feature values so that they are on a similar scale or range. This is important because features with very different scales can bias many machine learning algorithms. Here are the key points:

Why Scale?

Features with larger values might dominate the learning process.
It improves the convergence speed of gradient-based algorithms.
It helps algorithms that compute distances (like k-NN or clustering) work better.
Common Methods:

Normalization:
Rescales the data to a fixed range, typically 0 to 1.

```python
from sklearn.preprocessing import StandardScaler
```

```python
feature_x=StandardScaler().fit_transform(x)
```

```python
feature_x
```

# Split Data-set

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(feature_x,y,test_size=0.25,random_state=42)
xtest
```

# Building Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
pred = model.predict(xtest)
```

```python
pred.shape
```

#Peformce Metrics

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,pred)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(ytest,pred)
```

#Classification report of Confusion Matrix

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,pred))
```
