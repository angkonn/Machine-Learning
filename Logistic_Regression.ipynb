```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('marital status.csv')
df.head(20)
```

```python
df.isnull().sum()
```

We will not find **mean** and drop the null values of the status column,
because, 1 means yes, 0 means no. There's nothing in between.
If it was the 'age column', we would definitely find the mean and place it in the null location.
So for now, we will find **median** for replacing the null places of the status column.

```python
handle = df['status'].median()
handle
```

median = 0 means between 0 and 1, 0 is appeared the most

```python
df.status = df.status.fillna(handle)
df
```

```python
df['status'].value_counts()
```

Here, age is the independent variable (x), and independent variables are put inside double brackets

```python
x = df[['age']]
x
```

```python
y = df['status']
y
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
xtest
```

```python
ytest
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(xtrain, ytrain)
```

```python
model.predict(xtest)
```

```python
model.score(xtest, ytest)
```
